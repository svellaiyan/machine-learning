```python
# Columns with most NULL values 
cutoff = 85
data_1 = data
def drop_missing_cols(cutoff):
    for column in data.columns:
        if sum(data[column].isnull()*1)*100/1460 > cutoff:
            print "Dropping",column, "; Total rows missing", sum(data[column].isnull()*1), sum(data[column].isnull()*1)*100/1460, "%"
            data_1.drop(columns=[column])

drop_missing_cols(cutoff)
```
